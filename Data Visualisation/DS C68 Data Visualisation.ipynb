## Boxplot

#### It is used to see the quartile wise distibution of numerical column
#### It is used to check the outliers
#### It is used to see the quartile wise relatioship between categrorical and numerical data

```python
df["windspeed"].describe()
```

```python
sns.boxplot(x=df["windspeed"])
plt.show()
#whiskers/handles
#quartile = Quarter + Percentile (25%,50%,75%)
# First (leftmost) side of box --> First Quartile (Q1-25%)
# Middle (Center) side of box --> Second Quartile (Q2-50%) (Median)
# Last (Rightmost) side of box --> Third Quartile (Q3-75%)
# Emperical Relationships
# Lower Whiskers = Q1-1.5*(Q3-Q1)
# Higher Whiskers = Q3+1.5*(Q3-Q1)
```

```python
uw=15.625589+1.5*(15.625589-9.041650)
df[df["windspeed"]>uw]["windspeed"].count()
```

```python
sns.boxplot(x=df["season"],y=df["temp"])
plt.xlabel("Seasons")
plt.ylabel("Temperature")
plt.title("Boxplot of temperature for different seasons")
plt.show()
```

# HeatMap

```python
# Correlation: Strength of relationship (-1 to 1)
# Positive Correlation: Positive Relationship (0 to 1)
# Negative Correlation: Negative Relationship (-1 to 0)
# No Correlation: 0
```

```python
>0.7 to 1 --> Strong Positive Correlation
>0.3 to 0.7 --> Weak Positive Correlation
<-0.7 to -1 --> Strong Negative Correlation
<-0.3 to -0.7 --> Weak Negative Correlation
>-0.3 to 0.3 --> No Correlation
```

```python
plt.figure(figsize=(8,5))
sns.heatmap(df[["temp","hum","windspeed","atemp","cnt",'casual']].corr(),annot=True)
plt.show()
```
